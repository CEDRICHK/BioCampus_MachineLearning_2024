# TP : Classification of Malaria Cell Images
## Objective

In this practical, you will learn how to create and train machine learning models to classify images of malaria-infected and uninfected cells.

## Step 1: Import necessary libraries
### 1.1: Import the required libraries

```python
import os
import numpy as np
import cv2
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import multiprocessing
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Define paths to the data directories
### 2.1: Verify and set the working directory

```python
# Check current working directory
print(os.getcwd())

# Set working directory if necessary
#os.chdir('./Documents/BioCampus_MDL_2024/exercises/python')
#print(os.getcwd())  # Verify the change
```

### 2.2: Specify the paths to the directories containing the images of infected and uninfected cells

```python
# Paths to the data directories
infected_dir = './malaria_dataset/cell_images/Parasitized/'
uninfected_dir = './malaria_dataset/cell_images/Uninfected/'
new_images_dir = './malaria_dataset/cell_images/new_images/'
```

## Step 3: Load and preprocess images
### 3.1: Create a function to load and preprocess images. This function will resize the images, convert them to grayscale, and add labels

```python
data = []
labels = []

# Number of images to use
num_images = 1000

# Function to load and preprocess images
def load_images(directory, label, num_images):
    images = []
    labels = []
    for img_name in os.listdir(directory)[:num_images]:
        img_path = os.path.join(directory, img_name)
        img = cv2.imread(img_path)
        if img is not None:  # Check if the image is loaded correctly
            img = cv2.resize(img, (64, 64))  # Resize the images to make them uniform
            #img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            images.append(img)
            labels.append(label)
    return images, labels
```

### 3.2: Load images of infected and uninfected cells

```python
# Load infected images
infected_images, infected_labels = load_images(infected_dir, 'Parasitized', num_images)
data.extend(infected_images)
labels.extend(infected_labels)

# Load uninfected images
uninfected_images, uninfected_labels = load_images(uninfected_dir, 'Uninfected', num_images)
data.extend(uninfected_images)
labels.extend(uninfected_labels)
```

## Step 4: Visualize the images
### 4.1: Visualize a few samples of the loaded images to understand the data better.

```python
# Visualize the first 10 infected and uninfected images
plt.figure(figsize=(10, 5))

# Plot infected images
for i in range(10):
    plt.subplot(2, 10, i + 1)
    plt.imshow(cv2.cvtColor(infected_images[i], cv2.COLOR_BGR2RGB))
    plt.title("Infected")
    plt.axis('off')

# Plot uninfected images
for i in range(10):
    plt.subplot(2, 10, 10 + i + 1)
    plt.imshow(cv2.cvtColor(uninfected_images[i], cv2.COLOR_BGR2RGB))
    plt.title("Uninfected")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Step 5: Prepare the data

### 5.1: Convert the data and labels lists to numpy arrays, flatten the images, and encode the labels

```python
# Convert data and labels to numpy arrays
data = np.array(data)
labels = np.array(labels)

# Flatten the images
data = data.reshape(data.shape[0], -1)

# Encode the labels
le = LabelEncoder()
labels = le.fit_transform(labels)
```

## Step 6: Split the data into training and test sets

### 6.1: Split the data into training and test sets

#### Task: Complete the following code to split the data

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
(X_train.shape[0], X_test.shape[0])
```

## Step 7: Create and train the SVC model

### 7.1: Create and train a machine learning model using a Support Vector Classifier (SVC)

#### Task: Complete the following code to create and train the SVC model

```python
# Create and train the SVC model
svc = SVC(kernel='linear')
svc.fit(X_train, y_train)

# Predict and evaluate the SVC model
# (fill in the blanks)
y_pred_svc = svc.predict(X_test)
svc_accuracy = accuracy_score(y_test, y_pred_svc)

print("SVC Accuracy (without optimization):", svc_accuracy)
```

## Step 8: Create and train the Random Forest model

### 8.1: Create and train a model using Random Forest

#### Task: Complete the following code to create and train the Random Forest model

```python
# Create and train the Random Forest model
rf = RandomForestClassifier(n_estimators=100, n_jobs=multiprocessing.cpu_count())
rf.fit(X_train, y_train)

# Predict and evaluate the Random Forest model
y_pred_rf = rf.predict(X_test)
rf_accuracy = accuracy_score(y_test, y_pred_rf)

print("Random Forest Accuracy (without optimization):", rf_accuracy)
```

## Step 9: Generate confusion matrix for the test set predictions

### 9.1: Generate confusion matrix and classification report for the test set predictions

#### Task: Complete the following code to generate and display the confusion matrix

```python
# Generate confusion matrix for SVC
conf_matrix_svc = confusion_matrix(y_test, y_pred_svc)
sns.heatmap(conf_matrix_svc, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Confusion Matrix - SVC")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Generate confusion matrix for Random Forest
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Print classification report for SVC
print("Classification Report - SVC")
print(classification_report(y_test, y_pred_svc))

# Print classification report for Random Forest
print("Classification Report - Random Forest")
print(classification_report(y_test, y_pred_rf))
```

## Step 10: Classify new images
### 10.1: Create a function to load and preprocess new images

```python
# Function to load new images without labels
def load_new_images(directory):
    images = []
    filenames = []
    for img_name in os.listdir(directory):
        img_path = os.path.join(directory, img_name)
        img = cv2.imread(img_path)
        if img is not None:
            img = cv2.resize(img, (64, 64))  # Resize the images to make them uniform
            #img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale (Optional, depending on your choice)
            images.append(img)
            filenames.append(img_name)
    return images, filenames
```

### 10.2: Load and preprocess new images for classification

```python
# Load new images
new_images, new_filenames = load_new_images(new_images_dir)

# Convert new images to numpy array and flatten them
new_images = np.array(new_images)
new_images = new_images.reshape(new_images.shape[0], -1)
```

### 10.3: Predict the labels for new images

#### Task: Complete the following code to predict the labels for new images using the trained models

```python
# Predict using the SVC model
new_predictions_svc = svc.predict(new_images)

# Predict using the Random Forest model
new_predictions_rf = rf.predict(new_images)
```

### 10.4: Visualize the predictions for the new images

```python
plt.figure(figsize=(25, 12))

for i in range(len(new_images)):
    plt.subplot(10, 10, i + 1)
    plt.imshow(cv2.cvtColor(new_images[i].reshape(64, 64, 3), cv2.COLOR_BGR2RGB))
    plt.title(f"Pred: {new_predictions_rf[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Step 11: Optimize hyperparameters with GridSearchCV

### 11.1: Use GridSearchCV to optimize the hyperparameters of the SVC model

#### Task: Complete the following code to use GridSearchCV for SVC

```python
# Parameters for GridSearchCV for SVC
svc_params = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.0001, 0.001, 0.1, 1],
    'kernel': ['linear', 'rbf', 'poly']
}

# Create and train the SVC model with GridSearchCV
svc = GridSearchCV(SVC(), svc_params, cv=10, n_jobs=multiprocessing.cpu_count())
svc.fit(X_train,y_train)

# Predict and evaluate the SVC model
y_pred_svc = svc.predict(X_test)
svc_accuracy = accuracy_score(y_test, y_pred_svc)

# Display the best parameters for SVC
print("Best parameters for SVC:", svc.best_params_)
print("SVC Accuracy (with optimization):", svc_accuracy)
```

### 10.2: Use GridSearchCV to optimize the hyperparameters of the Random Forest model

#### Task: Complete the following code to use GridSearchCV for Random Forest

```python
# Parameters for GridSearchCV for Random Forest
rf_params = {
    'n_estimators': [50, 100, 200, 300, 400],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 15, 20],
    'min_samples_leaf': [1, 2, 4, 6],
    'bootstrap': [True, False]
}

# Create and train the Random Forest model with GridSearchCV
rf = GridSearchCV(RandomForestClassifier(n_jobs=multiprocessing.cpu_count()), rf_params, cv=10, n_jobs=multiprocessing.cpu_count())
rf.fit(X_train, y_train)

# Predict and evaluate the Random Forest model
y_pred_rf = rf.predict(X_test)
rf_accuracy = accuracy_score(y_test, y_pred_rf)

# Display the best parameters for Random Forest
print("Best parameters for Random Forest:", rf.best_params_)
print("Random Forest Accuracy (with optimization):", rf_accuracy)
```
