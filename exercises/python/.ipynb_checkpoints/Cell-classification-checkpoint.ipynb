## Step 3: Load and preprocess images
### 3.1: Create a function to load and preprocess images. This function will resize the images, convert them to grayscale, and add labels

```python
data = []
labels = []

# Number of images to use
num_images = 100

# Function to load and preprocess images
def load_images(directory, label, num_images):
    images = []
    labels = []
    for img_name in os.listdir(directory)[:num_images]:
        img_path = os.path.join(directory, img_name)
        img = cv2.imread(img_path)
        if img is not None:  # Check if the image is loaded correctly
            img = cv2.resize(img, (64, 64))  # Resize the images to make them uniform
            images.append(img)
            labels.append(label)
    return images, labels
```

### 3.2: Load images of infected and uninfected cells

```python
# Load infected images
infected_images, infected_labels = load_images(infected_dir, 'Parasitized', num_images)
data.extend(infected_images)
labels.extend(infected_labels)

# Load uninfected images
uninfected_images, uninfected_labels = load_images(uninfected_dir, 'Uninfected', num_images)
data.extend(uninfected_images)
labels.extend(uninfected_labels)
```
